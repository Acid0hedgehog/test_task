```python
#создадим 2 фунции для получения курса валют
import xml.etree.ElementTree as ET
import requests
import datetime

def get_exchange_rate(date):
    # создаем адрес
    url = f'http://www.cbr.ru/scripts/XML_daily.asp?date_req={date.strftime("%d/%m/%Y")}'
    
    response = requests.get(url)
    
    # Проверяем успешно ли прошел запрос
    if response.status_code == 200:
        xml_data = response.content.decode('cp1251')  # конверируем в строку
        return parse_exchange_rate(xml_data)
    else:
        # ошибка
        print(f"Failed to fetch data. Status code: {response.status_code}")
        return None

def parse_exchange_rate(xml_data):
    #Получаем корневой элемент
    root = ET.fromstring(xml_data)
    
    usd_rate = None
    eur_rate = None

    for valute in root.findall(".//Valute"):
        char_code = valute.find("CharCode").text
        value = valute.find("Value").text.replace(',', '.')  # Заменяем запятую на точку для корректного преобразования в число

        if char_code == "USD":
            usd_rate = float(value)
        elif char_code == "EUR":
            eur_rate = float(value)

    return usd_rate, eur_rate
```

```python
# Фильтрация строк по значению 'currency'
filtered_df = df1[df1['currency'].isin(['USD', 'EUR'])]
filtered_df
```

```python
# Создаем два новых столбца, в которые записываем курсы валют в даты из столбца 'created'
filtered_df.loc[:, 'usd_rate'] = None
filtered_df.loc[:, 'eur_rate'] = None

for i in range(len(filtered_df)):
    date_to_check = filtered_df['created'].iloc[i]
    rates = get_exchange_rate(date_to_check)

    if rates is not None:
        usd_rate, eur_rate = rates
        filtered_df.loc[:, 'usd_rate'].iloc[i] = usd_rate
        filtered_df.loc[:, 'eur_rate'].iloc[i] = eur_rate
filtered_df
```

```python
# Умножаем значения salary_from и salary_to на курсы валют
usd_filter = filtered_df['currency'] == 'USD'
eur_filter = filtered_df['currency'] == 'EUR'

filtered_df.loc[usd_filter, 'salary_from'] = (filtered_df.loc[usd_filter, 'salary_from'] * filtered_df.loc[usd_filter, 'usd_rate']).apply(lambda x: round(x, 2))
filtered_df.loc[usd_filter, 'salary_to'] = (filtered_df.loc[usd_filter, 'salary_to'] * filtered_df.loc[usd_filter, 'usd_rate']).apply(lambda x: round(x, 2))

filtered_df.loc[eur_filter, 'salary_from'] = (filtered_df.loc[eur_filter, 'salary_from'] * filtered_df.loc[eur_filter, 'eur_rate']).apply(lambda x: round(x, 2))
filtered_df.loc[eur_filter, 'salary_to'] = (filtered_df.loc[eur_filter, 'salary_to'] * filtered_df.loc[eur_filter, 'eur_rate']).apply(lambda x: round(x, 2))

filtered_df
```

```python
# Удаляем столбцы
filtered_df = filtered_df.drop(['usd_rate', 'eur_rate'], axis=1)
filtered_df
```

#### 3. Сделайте рейтинг (топ-30) навыков по всем вакансиям.

```python
# Создадим датасет и проверим ти данных в столбце 'skills'
df3 = df1.copy(deep=True)
type(df3['skills'][0])
```

```python
# Создаем словарь, для подсчета количества встреч навыков
skills = {}

# Проходим по всем строкам столбца 'skills'
for index, row in df3.iterrows():
    # Проверям,не является ли строка пустой и рабливаем ее по ';'
    if pd.notna(row['skills']):
        skills_row = row['skills'].split(';')
        
        for skill in skills_row:
            skill = skill.strip()  # Убираем лишние пробелы
            # Записываем значение в словарь, если его там нет или увеличиваем на 1 количество встреч
            if skill not in skills:
                skills[skill] = 1
            else:
                skills[skill] += 1

# В строках часто присутствует пробел в конце строки, это нужно исправить
if '' in skills:
    del skills['']
    
# Сортируем    
sorted_skills = sorted(skills.items(), key=lambda x: x[1], reverse=True)

# Выводим топ-30 навыков
top_30_skills = sorted_skills[:30]
print('Top 30 Skills:')
for skill, count in top_30_skills:
    print(f'{skill}: {count}')
```

#### 4. Cоставьте топ-10 городов с наибольшей средней заработной платой по вакансиям тестировщика (QA).

```python
df4 = df1.copy(deep=True)
df4
```

```python
# Найдем вакансии где есть QA
df_qa = df4[df4['vacancy'].str.contains('QA', case=False, na=False)]
df_qa
```

```python
# Убираем вакансии где нет заработной платы
df_qa = df_qa.dropna(subset=['salary_from', 'salary_to'], how='all')
df_qa
```

```python
# Создаем новый столбец, по следующим критериям
# если в столбце salary_from стоит NaN, то записать значение столбца salary_to
# если в столбце salary_toстоит NaN, то записать значение столбца salary_from 
# если в обоих столбцах есть значение, то записать их среднее арифметическое
df_qa['avg_salary'] = df_qa.apply(lambda row: row['salary_to'] if pd.isna(row['salary_from'])
                                          else row['salary_from'] if pd.isna(row['salary_to'])
                                          else (row['salary_from'] + row['salary_to']) / 2, axis=1)
df_qa
```

```python
# Создаем таблицу с городами и средней заработной платой
avg_salary_by_area = df_qa.groupby('area')['avg_salary'].mean().reset_index()
avg_salary_by_area['avg_salary'] = avg_salary_by_area['avg_salary'].round(2)
avg_salary_by_area
```

```python
import pandas as pd

def search_vacancies(df, min_salary=None, max_salary=None, has_test=None, keyword=None, skill=None):
    # Фильтрация по зарплате
    if min_salary is not None:
        df = df[df['salary_from'] >= min_salary]

    # Фильтрация по наличию тестового задания
    if has_test is not None:
        df = df[df['has_test'] == has_test]

    # Фильтрация по наличию ключевого слова в наименовании вакансии
    if keyword is not None:
        df = df[df['vacancy'].str.contains(keyword, case=False)]

    # Обработка NaN значений в 'skills' перед фильтрацией
    df['skills'] = df['skills'].fillna('')  

    # Фильтрация по наличию определенного навыка
    if skill is not None:
        df = df[df['skills'].str.contains(skill, case=False)]

    # Вывод результатов
    num_found = len(df)
    avg_salary = df[['salary_from', 'salary_to']].mean().mean().round(2)
    found_urls = df['url'].tolist()

    print(f"Найдено вакансий: {num_found}")
    print(f"Средняя зарплата: {avg_salary}")
    print("Ссылки на вакансии:")
    for url in found_urls:
        print(url)

# Пример использования
search_vacancies(df5, has_test=False, keyword='Senior', skill='Java')
search_vacancies(df5, keyword='Senior',)
```
